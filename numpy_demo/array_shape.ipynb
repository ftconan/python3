### 数组形状操作

```python
import numpy as np
tang_array = np.arange(10)
tang_array
```

```python
tang_array.shape
```

```python
tang_array.shape = 2,5
```

```python
tang_array
```

数组大小不能改变

```python
tang_array.reshape(1,10)
```

```python
tang_array.shape = 3,4
```

```python
tang_array = np.arange(10)
tang_array.shape
```

```python
tang_array = tang_array[np.newaxis,:]
tang_array.shape
```

```python
tang_array = tang_array[:,np.newaxis]
tang_array.shape
```

```python
tang_array = tang_array[:,np.newaxis,np.newaxis]
tang_array.shape
```

```python
tang_array = tang_array.squeeze()
tang_array.shape
```

```python
tang_array.shape = 2,5
```

```python
tang_array
```

```python
tang_array.transpose()
```

```python
tang_array.T
```

```python
tang_array
```

数组的连接

```python
a = np.array([[123,456,678], [3214,456,134]])
a
```

```python
b = np.array([[1235,3124,432], [43,13,134]])
b
```

```python
c = np.concatenate((a,b))
c
```

```python
c = np.concatenate((a,b), axis=0)
c
```

```python
c = np.concatenate((a,b), axis=1)
c
```

```python
c.shape
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```
